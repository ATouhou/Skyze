# Check Exchange Data History Provided

```python
start_date = datetime(2010, 1, 1)
interval = "1d"
exclude = ["_1broker"]

start_date_unix_ts = int(p_start_date.timestamp()*1000)


for ident in ccxt.exchanges:
    print("\nExchange: ",ident)
    exchange = eval ('ccxt.%s ()' % ident)

    if exchange.hasFetchOHLCV and ident not in exclude:
        # Get markets and select first
        market = next(iter(exchange.load_markets())) # Type dict of dict {Mkt_Name: {Mkt_deets}}
        
        
        print("Market: ",market)
        try:
            market_history = exchange.fetch_ohlcv(market, interval,
                                        since = start_date_unix_ts)

            print("Since: ", start_date_unix_ts)
            print("Rows in Market History: ",len(market_history))
            if len(market_history)>0:
                print("First row: ", exchange.iso8601(market_history[1][0]), market_history[1])
                print("Last row: ", exchange.iso8601(market_history[-1][0]), market_history[-1])
        except:
            print("exception")
```
