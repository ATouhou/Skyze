# Map function testing

```python
import numpy as np
import pandas as pd
def calc_signal(self, p_bullish, p_bearish):
    signal = 0
    if p_bullish: signal = 1
    elif p_bearish: signal = -1
    return bool(signal)
```

```python
p_data = pd.DataFrame(np.random.randint(0,2,size=(100, 2)), columns=list(['Bullish','Bearish']))
print(p_data.head(10))
```

```python
def f(x):  
    result = 0
    if (x[0]) == 1 : result = 1
    elif (x[1]) == 1 : result = -1
    return  result
```

```python
p_data["Signal"] = p_data[['Bullish','Bearish']].apply(f, axis=1) #passes a Series object, row-wise
```
